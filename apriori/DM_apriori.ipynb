```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

data = pd.read_excel('shopping.xlsx')
pd.DataFrame(data)
data.head(10)
```

```python
data1 = apriori(data,min_support=0.2,use_colnames=True)
data1.head(20)
```

```python
data2 = association_rules(data1,metric='confidence',min_threshold=0.55)
data2.head(10)
```
